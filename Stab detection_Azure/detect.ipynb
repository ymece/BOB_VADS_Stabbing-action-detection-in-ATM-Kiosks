```python
from predict import detect_image
from tkinter import *
import cv2
import sys
from PIL import ImageTk, Image
import winsound
from tkvideo import tkvideo
import tkinter as tk
from tkinter import ttk
from tkinter import filedialog as fd
import cv2
import sys
import numpy as np
import glob
import matplotlib.pyplot as plt
%matplotlib inline
from tkinter.messagebox import showinfo
root = tk.Tk()
root.title('Stabbing Action Detection in ATM kiosks')
w = tk.Label(root, text="STABBING ACTION DETECTION AT ATM KIOSKS",font=("Arial", 18))
w.pack()
root.geometry('700x350')
root.resizable(width=True, height=True)
app = Frame(root, bg="white")
label =Label(root)
label.pack()
global a
global l


colors = {
        "knifeinhand": (0,215,255)
}


def select_file():
    filename=fd.askopenfilename(filetypes=(("Video files", "*.mp4"),))

    showinfo(
        title='Selected File',
        message=filename
        
    )
    return filename


def run():
    #camera=fd.askopenfilename(filetypes=(("Images", "*.jpg"),))
    #a = cv2.imread(camera) 
    x=select_file()
  
    video = cv2.VideoCapture(x)
    n = 0
    while True:
    
        ret, a = video.read()
        if not ret:            break
        cv2.imwrite('input_frames/capture%04i.jpg'%n, a)
        results = detect_image('input_frames/capture%04i.jpg'%n)
        l = 0
        print('frame_sent')
        n = n+1
        img_width = 2400
        img_height = 1080
        
        for prediction in results:
            if prediction['probability'] > 0.1:
                print(f"{prediction['tagName']}: {prediction['probability'] * 100 :.2f}%")
                color = colors[prediction['tagName']]
                print('2')
                left = prediction['boundingBox']['left'] * img_width
                top = prediction['boundingBox']['top'] * img_height
                height = prediction['boundingBox']['height'] * img_height
                width =  prediction['boundingBox']['width'] * img_width
                result_image = cv2.rectangle(a, (int(left), int(top)), (int(left + width), int(top + height)), color, 3)
                cv2.putText(result_image, f"{prediction['tagName']}: {prediction['probability'] * 100 :.2f}%", (int(left), int(top)-10), fontFace = cv2.FONT_HERSHEY_SIMPLEX, fontScale = 0.5, color = color, thickness = 2)
                #stacked = np.hstack((a, result_image))
                #cv2.imshow('Knife detected',result_image)
            
                l = l + 1
                cv2.imwrite('output_frames/result%04i.jpg'%n, result_image)
                print('3')
                frequency=2000
                duration=1500
                winsound.Beep(frequency,duration)
                
            

    
    #camera.set(cv2.CAP_PROP_FRAME_WIDTH, img_width)
    #camera.set(cv2.CAP_PROP_FRAME_HEIGHT, img_height)

    #ret, image = camera.read()
    #cv2.imwrite('capture.jpg', a)

    #results = detect_image('capture/0i.jpg'%n)
    print("\n", "*"*8, "Knife detected in ATM!", "*"*8, "\n")
    print("\n", "*"*8, "Resultant frames are stored in Output folder", "*"*8, "\n")
    print("\n", "*"*8, "The resultant video gets saved once the GUI is closed", "*"*8, "\n")

# Select color for the bounding box (BGR)
         

gmm_button=Button(root,text='Microsoft Azure',command=run).pack()



# Close the windows.q
cv2.destroyAllWindows()

root.mainloop()

img_array = []
for filename in glob.glob('output_frames/*.jpg'):
    img = cv2.imread(filename)
    height, width, layers = img.shape
    size = (width,height)
    img_array.append(img)


    out = cv2.VideoWriter('Knife_detected_video.mp4',cv2.VideoWriter_fourcc(*'DIVX'), 15, size)


#cv2.VideoWriter("/your/path/filename.mp4",fourcc, 20,(320,180))
    print('Knife Detected Video saved')
 
    for i in range(len(img_array)):
        out.write(img_array[i])
out.release()
```
